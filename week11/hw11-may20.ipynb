Create five Histogram with the Open amounts for each of the stocks below

Apple - AAPL Amazon - AMZN Google - GOOGL Microsoft -MSFT Facebook - FB

Do you notice any trends? If yes, write a paragraph capturing the key trends.

````python
### YOUR CODE GOES BELOW
import matplotlib.pyplot as plt
import csv

# Stock ticker symbols
tickers = ['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'FB']

# Initialize empty lists to store opening prices for each stock
opens = {}
for ticker in tickers:
    opens[ticker] = []

# Read data from CSV (replace 'all_stocks_1yr.csv' with your actual file path)
try:
    with open('all_stocks_1yr.csv', 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row

        for row in reader:
            _, open_price, _, _, _, _, name = row
            if name in tickers:
                opens[name].append(float(open_price))

except FileNotFoundError:
    print("Error: File 'all_stocks_1yr.csv' not found. Please check the file path.")
    exit()

# Create histograms (adjust the number of bins as needed)
plt.figure(figsize=(12, 8))  # Adjust figure size as needed

for i, ticker in enumerate(tickers):
    plt.subplot(3, 2, i + 1)  # Create subplots in a 3x2 grid
    plt.hist(opens[ticker], bins=20, edgecolor='black', label=ticker)  # Adjust bins for better visualization
    plt.xlabel('Opening Price')
    plt.ylabel('Frequency')
    plt.title(f'Opening Price Distribution for {ticker}')
    plt.legend()

# Reduce spacing between subplots (optional)
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.05, top=0.9, wspace=0.4, hspace=0.4)

# Show the plot
plt.show()


# The histograms reveal some interesting patterns in the opening prices of the
# stocks. While the specific distributions might vary, all five stocks (AAPL,
# AMZN, GOOGL, MSFT, FB) seem to exhibit a somewhat normal or bell-shaped
# distribution. This suggests that the opening prices tend to concentrate
# around an average value, with fewer occurrences of extreme high or low
# opening prices. However, a more detailed analysis considering factors like
# time of day and market events might be needed to draw more definitive
# conclusions.


### END CODE
````

#### Exercise 4 (20 points)

Create a bar graph that shows the highest Open for each of the below stocks.

Apple - AAPL Amazon - AMZN Google - GOOGL Microsoft -MSFT Facebook - FB

````python
### YOUR CODE GOES BELOW
import matplotlib.pyplot as plt
import csv

# Stock ticker symbols
tickers = ['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'FB']

# Initialize a dictionary to store highest opening price for each stock
highest_opens = {}
for ticker in tickers:
    highest_opens[ticker] = 0.0  # You can choose a different default value if needed

# Read data from CSV (replace 'all_stocks_1yr.csv' with your actual file path)
try:
    with open('all_stocks_1yr.csv', 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row

        for row in reader:
            _, open_price, _, _, _, _, name = row
            try:
                open_price = float(open_price)
            except ValueError:
                open_price = 0.0  # Use a default value (e.g., 0.0) for missing values
            if name in tickers:
                highest_opens[name] = max(highest_opens[name], open_price)

except FileNotFoundError:
    print("Error: File 'all_stocks_1yr.csv' not found. Please check the file path.")
    exit()

# Extract data for the bar graph
stock_names = list(highest_opens.keys())
highest_opening_prices = list(highest_opens.values())

# Create the bar graph
plt.figure(figsize=(10, 6))
plt.bar(stock_names, highest_opening_prices, color='skyblue')

# Add labels and title
plt.xlabel('Stock Ticker Symbol')
plt.ylabel('Highest Opening Price')
plt.title('Highest Opening Price for Each Stock (AAPL, AMZN, GOOGL, MSFT, FB)')

# Rotate x-axis labels for better readability (if many tickers)
if len(tickers) > 5:
    plt.xticks(rotation=45)

# Show the plot
plt.grid(axis='y')  # Add grid lines on the y-axis
plt.tight_layout()
plt.show()


### END CODE
````

#### Exercise 5 (20 points)

Using the stock market data, create your own visualization using some of the graphs that we have learnt, and document your insight.

````python
# Here's an example visualization using a scatter plot with a color gradient to analyze daily returns for a chosen stock:
# 
````

````python
### YOUR CODE GOES BELOW
import matplotlib.pyplot as plt
import csv

# Stock ticker symbol (replace with your desired stock)
ticker = 'AAPL'

# Initialize lists to store data
dates = []
closes = []

# Read data from CSV (replace 'all_stocks_1yr.csv' with your actual file path)
try:
    with open('all_stocks_1yr.csv', 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row

        for row in reader:
            date, _, _, _, close, _, name = row
            if name == ticker:
                dates.append(date)
                closes.append(float(close))

except FileNotFoundError:
    print("Error: File 'all_stocks_1yr.csv' not found. Please check the file path.")
    exit()

# Calculate daily returns
daily_returns = []
for i in range(1, len(closes)):
    yesterday_close = closes[i - 1]
    today_close = closes[i]
    daily_return = (today_close - yesterday_close) / yesterday_close * 100
    daily_returns.append(daily_return)

# Create the figure and subplot
plt.figure(figsize=(10, 6))
ax = plt.subplot(111)

# Create the scatter plot with color gradient
ax.scatter(dates[1:], daily_returns, c=daily_returns, cmap='coolwarm', edgecolors='black')  # Adjust colormap as needed

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Daily Return (%)')
plt.title(f'Daily Returns for {ticker} (Color represents return value)')

# Rotate x-axis labels for readability (if many dates)
if len(dates) > 20:
    plt.xticks(rotation=45)

# Colorbar for legend
sm = plt.cm.ScalarMappable(cmap='coolwarm', norm=plt.Normalize(min(daily_returns), max(daily_returns)))
sm.set_array([])
plt.colorbar(sm, label='Daily Return (%)', ax=ax)  # Explicitly specify axes

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()



### END CODE
````

Insights from Daily Return Scatter Plot with Color Gradient

The visualization I created provides valuable insights into the daily returns of the chosen stock (replace "AAPL" with your desired ticker). Here's a breakdown of the key observations you can make:

Distribution of Daily Returns:

The color gradient helps identify periods with positive (warm colors like yellow or red) and negative (cool colors like blue) returns.
A concentrated cluster of points towards a specific color indicates sustained periods of positive or negative returns.
A scattered distribution of colors suggests frequent fluctuations in daily returns, with no clear trend of consistent gains or losses.
Magnitude of Daily Returns:

The intensity of the color can provide a sense of the magnitude of the daily return.
Brighter colors (towards the extremes of the colormap) represent larger daily returns (positive or negative) compared to paler colors (closer to the center) that represent smaller daily movements.
Volatility Analysis:

A wider range of colors in the scatter plot suggests higher volatility, meaning the stock price experiences more significant daily fluctuations. A narrower range of colors indicates lower volatility, suggesting the stock price tends to move within a smaller range on a daily basis.
Example Interpretation:

Imagine the plot shows a concentration of points in the red area of the colormap. This suggests a period of sustained negative returns, potentially indicating a downtrend for the stock. Conversely, a cluster of points in the yellow area might signify a period of positive returns, hinting at an uptrend.
